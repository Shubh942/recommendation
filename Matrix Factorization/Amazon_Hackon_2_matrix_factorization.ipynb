```python
movies = movies[['id', 'original_title', 'genres']]
movies['movieId']=movies['id']
movies.drop(columns=['id'])
```

```python
ratings['movieId'] = ratings['movieId'].astype(str)
# movies['movieId'] = [float(str(i).replace(",", "")) for i in movies['movieId']]
# for i in range(len(ratings['movieId'])):
#   ratings['movieId'][i] = str(ratings['movieId'][i])
```

```python
combine_movie_rating = pd.merge(ratings, movies, on='movieId')
combine_movie_rating.head(10)
```

```python
columns = ['timestamp', 'genres','id']
combine_movie_rating = combine_movie_rating.drop(columns, axis=1)
combine_movie_rating.head(10)
```

```python
combine_movie_rating = combine_movie_rating.dropna(axis = 0, subset = ['original_title'])

movie_ratingCount = (combine_movie_rating.
     groupby(by = ['original_title'])['rating'].
     count().
     reset_index().
     rename(columns = {'rating': 'totalRatingCount'})
     [['original_title', 'totalRatingCount']]
    )
movie_ratingCount.head(10)
```

```python
rating_with_totalRatingCount = combine_movie_rating.merge(movie_ratingCount, left_on = 'original_title', right_on = 'original_title', how = 'left')
rating_with_totalRatingCount.head(10)
```

```python
user_rating = rating_with_totalRatingCount.drop_duplicates(['userId','original_title'])
user_rating.head(10)
```

```python
sampled_user_rating = user_rating.sample(frac=0.0001)
```

```python
sampled_user_rating
```

```python
movie_user_rating_pivot = sampled_user_rating.pivot(index='userId', columns='original_title', values='rating').fillna(0)
movie_user_rating_pivot.head(10)
```

```python
X = movie_user_rating_pivot.values.T
X
```

```python
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=12, random_state=17)
matrix = SVD.fit_transform(X)
matrix
```

```python
import warnings
warnings.filterwarnings("ignore",category =RuntimeWarning)
corr = np.corrcoef(matrix)
corr
```

```python
movie_title = movie_user_rating_pivot.columns
movie_title_list = list(movie_title)
coffey_hands = movie_title_list.index("Ober")
```

```python
corr_coffey_hands  = corr[coffey_hands]
list(movie_title[(corr_coffey_hands >= 0.9)])
```
